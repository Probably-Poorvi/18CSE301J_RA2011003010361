```python
import pandas as pd
```

```python
df = pd.read_csv('Mental_H.csv', low_memory=False)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(x='Year', y='Depression (%)', kind='line', title='Depression every year')
plt.show()
```

```python
df['Year'].hist(bins=20, color='blue', alpha=0.5)
plt.xlabel('Schizophrenia (%)')
plt.ylabel('Entity')
plt.title('Schizophrenia around the world')
plt.show()
```

```python
df.boxplot(column='Depression (%)', by='Entity')
plt.xlabel('Entity')
plt.ylabel('Depression (%)')
plt.title('Boxplot of Depression (%) by Entity')
plt.show()
```
